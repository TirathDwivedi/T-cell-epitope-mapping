```python
from matplotlib_venn import venn2, venn2_circles
import matplotlib.pyplot as plt

# Define the sizes of the sets and their overlap
v1_unique = 63
v2_unique = 12
common_count = 16

# Create a Venn diagram
venn = venn2(subsets=(v1_unique, v2_unique, common_count),
             set_labels=('BKV', 'JCV'), set_colors=('lightsteelblue', 'orange'), alpha=0.7)
venn2_circles(subsets=(v1_unique, v2_unique, common_count),
              linestyle='dashed', linewidth=2, color='black')

plt.show()
```
