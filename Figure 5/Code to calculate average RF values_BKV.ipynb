```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Load data from Excel file
df = pd.read_excel("BKV Large T_average RF values.xlsx")

# Load full protein sequence
with open("BKV Large T.fa", "r") as f:
    f.readline()  # Skip header line
    protein_seq = f.read().replace("\n", "")

# Create dictionary of RF values for each position
rf_dict = {}
for i, row in df.iterrows():
    seq = row["Sequence"]
    start_pos = row["Mapped Start Position"]
    end_pos = row["Mapped End Position"]
    rf = row["Response Freq."]
    for j in range(start_pos-1, end_pos):
        rf_dict[j] = max(rf_dict.get(j, 0), rf)

# Plot the RF values
fig, ax = plt.subplots(figsize=(8, 4))
x = range(len(protein_seq))
ax.plot(x, [rf_dict.get(i, 0) for i in range(len(protein_seq))], linewidth=1)
ax.set_xlabel("Position in protein", fontsize=16)
ax.set_ylabel("Response frequency", fontsize=16)
ax.tick_params(axis="both", which="major", labelsize=16, width=1.5, length=4)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.set_xticks(range(0, len(protein_seq), 200))
ax.set_xticklabels([str(i) for i in range(0, len(protein_seq), 200)])
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_linewidth(0.5)
ax.spines["left"].set_linewidth(0.5)
plt.tight_layout()
plt.show()
```
