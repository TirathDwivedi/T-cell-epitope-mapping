```python
import pandas as pd
import matplotlib.pyplot as plt

# read in the data file (replace with your own data file)
df = pd.read_excel('HLA supertypes_BKV.xlsx')

# create a horizontal bar chart
fig, ax = plt.subplots(figsize=(4, 3))

# plot the bars
bars = ax.barh(df['HLA supertype'], df['Count'], color='#7EC0EE')

# set the x-axis label
ax.set_xlabel('Count', fontsize=12)

# set the font size of the tick labels
ax.tick_params(axis='both', which='major', labelsize=12)

# set the font size of the y-axis labels
ax.tick_params(axis='y', labelsize=10)

# set the x-axis limit
ax.set_xlim([0, 40])

# display the value of each bar
for i, bar in enumerate(bars):
    ax.annotate(str(df['Count'].iloc[i]), xy=(bar.get_width(), bar.get_y() + bar.get_height() / 2),
                xytext=(5, 0), textcoords='offset points', va='center', fontsize=12)

# remove the spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# remove padding
plt.tight_layout(pad=1)

# display the plot
plt.show()
```
